#### temperatura

**Jak działa temperatura:**
- **Niska temperatura (0.0-0.3)**: Model wybiera najbardziej prawdopodobne tokeny → bardziej deterministyczne, przewidywalne odpowiedzi
- **Średnia temperatura (0.7-1.0)**: Model wybiera z szerszego zakresu prawdopodobnych tokenów → bardziej zbalansowane odpowiedzi
- **Wysoka temperatura (1.0-2.0)**: Model wybiera z bardzo szerokiego zakresu → bardziej kreatywne, zaskakujące, ale czasem mniej spójne odpowiedzi

Wartość domyślna to zazwyczaj 1.0.


```python
response = client.completions.create(
    model="gpt-3.5-turbo-instruct",
    prompt="Finish the sentence: 'Litwo, Ojczyzno moja'",
    temperature=2.0,
    max_tokens=500
)
response.choices[0].text
```

#### generowanie wiecej nic jednej odp batch
