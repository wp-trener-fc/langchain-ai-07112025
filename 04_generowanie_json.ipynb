```python
from utils import Settings
from openai import OpenAI
import json

api_key = Settings().openai_api_key
client = OpenAI(api_key=api_key)
```

#### json mode

- Otrzymujemy **tekst w formacie JSON** (string), a nie gotowy obiekt Pythonowy
- Musimy sami sparsować odpowiedź używając `json.loads()` lub podobnej metody
- Słowo "json" musi być w promptcie (może być samo słowo bez opisu, API nie sprawdza nic więcej)

```python
with open('data/emails/5e0d726e-4ff1-47f5-b02a-84e060aea8fe.txt', encoding='utf-8') as file:
    content = file.read()

PROMPT = """
Uzyskaj z tekstu informacje o numerze umowy, którego dotyczy zgłoszenie, skórcony opis zdarzenia oraz adres.
Zwróć obiekt json w nastepującym formacie:

{
    "contract_number": "contract_num",
    "short_desc": "desc",
    "address": {
        "city": "city_name",
        "postal_code": "postal_code"
        "street_name": "street"
        "house_number": "nr"
    }
}
Jesli czegoś nie zjandziesz w tekscie to w odpowiedznim polu zostaw pustą wartość
"""
```

```python
response = client.chat.completions.create(
    model='gpt-4o-mini',
    response_format={"type": "json_object"},
    messages=[
        {"role": "system", "content": PROMPT},
        {"role": "user", "content": content}
    ]
)

print(response.choices[0].message.content)
```

```python
json_str = response.choices[0].message.content
json.loads(json_str)
```
